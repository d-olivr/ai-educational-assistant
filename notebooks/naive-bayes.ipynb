# Algoritmo Naive Bayes

## Conceitual

Esse algoritmo classificador utiliza o **Teorema de Bayes** para inferir classes desconhecidas.

É chamado de naive/ingênuo pois, em sua abordagem, mesmo que algumas features sejam interdependentes, elas serão vistas de forma independente pelo algoritmo.

### Tipos de Naive Bayes

*   Gaussian Naive Bayes: Tipo que usaremos nesse caso de estudo. O Gaussian é indicado quando as variáveis são contínuas, em que os dados assumem uma distribuição normal.
*   Multinomial Naive Bayes: Muito utilizado para NLP (Processamento de LInguagem Natural) pois seu uso é indicado quando as features representam contagens/frequências. Não se aplica ao nosso caso.
*   Bernoulli Naive Bayes: utilizado para variáveis são discretas, podendo assumir apenas 2 valores possíveis. Poderia ser o nosso caso, levando em consideração que inciamos com uma previsão binária (0 = Reprovado e 1 = Aprovado). Mas temos o objetivo de testar a previsão com mais classificações, sendo assim optamos seguir com o Gaussian.


## Teorema de Bayes

### Fórmula
$$
P(A \mid B) = \frac{P(B \mid A) \, P(A)}{P(B)}
$$


### Legenda
- P(A | B) = é a inferência que procuramos. É a probabilidade do evento A ter acontecido, dado que o evento B ocorreu. É a crença atualizada na probabilidade do evento

- P(B | A) =  verossimilhança -> é a probabilidade de observarmos a evidência B, dado que a hipótese A é verdadeira. É o quão provável é a evidência sob a hipótese.

- P(A) = probabilidade a priori de \(A\) -> é a crença inicial sobre a probabilidade do evento A, antes de sabermos qualquer dado sobre o evento B.

- P(B) = probabilidade de \(B\) -> probabilidade de evidência, ou seja, a probabilidade total do evento B ocorrer sob todas as circunstâncias possíveis.

## Algoritmo

### Carregamento de bibliotecas e dados

```python
#importar bibliotecas
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#carregamento de dados
df = pd.read_csv('/content/2periodo.csv')
```

### Definição

Criação da coluna aprovado (1) e reprovado (0) e definição de variáveis independentes e **alvo** (qual matéria iremos inferir)

```python
df['status'] = (df['Engenharia de Software'] >= 5).astype(int) # boolean p/ verificar se é aprovado ou reprovado

X = df.drop(columns=['matricula', 'Engenharia de Software', 'status'])  # colunas que não utilizaremos no processamento
y = df['status']
```

```python
X = X.fillna(0) # boa prática p/tratamento de valores null, mas já fizemos a limpeza
```

### Treino e teste

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# divisão de porcentagem de treino e teste -> x = features, y = target
```

X_train = features para treinar o modelo (70%)

X_test = features para testar o modelo (30%)

y_train = respostas corretas do treino (70%)

y_test = respostas corretas do teste (30%)

Treinamento

```python
modelo = GaussianNB()
modelo.fit(X_train, y_train) #y_train é o "gabarito"
```

Previsão

```python
y_pred = modelo.predict(X_test)
```

### Resultados e análise

Métricas que utilizamos:
*  Acurácia total: previsões corretas dividida pela quantidade total de previsões. $\frac{VP+VN}{(VP+FVN+FP+FN)}$

* Precisão: verifica se o aluno realmente foi aprovado, quando o modelo infere que foi. $\frac{VP}{(VP+FP)}$

* Recall: verifica quantos dos que realmente foram aprovados, quantos o modelo encontrou. $\frac{VP}{(VP+FN)}$

* F1 Score: Faz uma **média harmônica** de precisão e recall. Se algum dos dois tiver um valor muito baixo quando comparado com o outro,  o valor do F1 score será punido.

* Support: Número de ocorrências reais de cada classe no conjunto de dados de teste

* Macro Average: Média **não ponderada** das métricas (Precisão, Recall, F1-Score) para cada classe independentemente, em seguida, tira a média. O que é igual ao peso a todas as classes, independente de quantas instâncias cada classe tem. $\frac{1}{C} \sum_{i=1}^{C} M_i$ onde $M_i$ é a métrica M para a classe  




* Weighted Average: Média **ponderada** das métricas (Precisão, Recall, F1-Score) para cada classe, em seguida, tira a média ponderando pelo "support" de cada classe. Útil quando se tem classes desbalanceadas, dando mais peso às classes com mais instâncias. $\sum_{i=1}^{C} M_i \times \frac{Support_i}{\sum_{j=1}^{C} Support_j}$ onde $M_i$ é a métrica M para a classe

Análises

```python
print("acurácia total:", accuracy_score(y_test, y_pred))

print("\n resultados:")
print(classification_report(y_test, y_pred))

print("\n matriz de confusão:")
print(confusion_matrix(y_test, y_pred))
```
